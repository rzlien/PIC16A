Broadcasting refers to the automatic expansion of one or more arrays to make a computation "make sense." Here's a simple example: 

```python
a = np.array([0, 1, 2])
b = np.ones((3, 3))
a.shape, b.shape
```
```
((3,), (3, 3))
```
```python
a+b
```
```
array([[1., 2., 3.],
       [1., 2., 3.],
       [1., 2., 3.]])
```

What has happened here is that the first array `a` has been "broadcast" from a 1d array into a 2d array of size 3x3 in order to match the dimensions of `b`. The broadcast version of `a` looks like this: 

```
array([[0., 1., 2.],
       [0., 1., 2.],
       [0., 1., 2.]])
```

Consult the notes above for many more examples of broadcasting. Pay special attention to the discussion of the [rules of broadcasting](https://jakevdp.github.io/PythonDataScienceHandbook/02.05-computation-on-arrays-broadcasting.html#Rules-of-Broadcasting).

### Your Solution

_b) Approach:_
The approach to solving this problem is reading the prodived material on unit testing and numpy array broadcasting. Then, copy the code into the appropriate function definition. Figure out the expected output based on the broadcasting rules and then create an assert statement for it. Wrap all 3 functions in a testing class that inherits from `unittest.TestCase`.

````python
# write your tester class here
import unittest

class TestBroadcastingRules(unittest.TestCase):
    """
    Class containing unit tests for numpy array broadcasting.
    Contains tests for each of the 3 rules. 
    """
    
    def test_rule_1(self):
        """
        According to Rule 1, "if the two arrays differ in their number of dimensions, the shape 
        of the one with fewer dimensions is padded with ones on its leading (left) side". a has
        a shape of (3,) while b has a shape of (1,3). Thus, the a has one dimension while b has 
        two. Since these dimensions do not match, the one with fewer dimensions (a in this case)
        is padded with a 1 on its left side, transforming it into a new shape of (1,3).
        
        After broadcasting, a and b now have the same shape and addition can occur entrywise. Thus,
        the final entry of c is the value of the final entry of a plus the final entry of b (1 + 2).
        -------------------------------------------------------------------------------------------
        
        Test for numpy array broadcasting Rule 1.
        "If the two arrays differ in their number of dimensions, the shape 
        of the one with fewer dimensions is padded with ones on its leading (left) side.""
        
        Checks if the array with fewer dimensions becomes padded with ones on left side. 
        Also checks if entry of sum has expected value.
        """
        a = np.ones(3)
        b = np.arange(3).reshape(1, 3) 
        c = a + b
        
        #check shape of c has expected output
        self.assertEqual(c.shape, (1, 3))
        
        #check final entry of C has expected output
        self.assertEqual(c[0,2], 3)
    
    
    def test_rule_2(self):
        """
        According to Rule 2, "if the shape of the two arrays does not match in any dimension, 
        the array with shape equal to 1 in that dimension is stretched to match the other shape".
        a has a shape of (1,3) while b has a shape of (3,3). These do not match, so a must be
        broadcasted to match b. a's new shape becomes (3,3). When a is broadcasted, it makes two
        new rows, with the same values in each column as the first row. 
        
        Since a and b are the same shape after a is broadcasted, the values can be added together
        entrywise. The value of c[1,2] is 6 because that is the sum of a[1,2] + b[1,2] = 1 + 5. 
        -------------------------------------------------------------------------------------------
        
        Test for numpy array broadcasting Rule 2.
        "If the shape of the two arrays does not match in any dimension, the array with shape 
        equal to 1 in that dimension is stretched to match the other shape."
        
        Checks if the array with shape equal to 1 is stretched to match the other array's. 
        Also checks if entry of sum has expected value.
        """
        a = np.ones((1, 3))
        b = np.arange(9).reshape(3, 3)
        c = a + b
        
        #check shape of c has expected output
        self.assertEqual(c.shape, (3,3))
        
        #check entry c[1,2] has expected value
        self.assertEqual(c[1,2], 6)
    
    def test_rule_3(self):
        """
        According to Rule 3, "if in any dimension the sizes disagree and neither is equal to 1, 
        an error is raised". a has a size of (2,3) and b has a size of (3,3). These dimensions do
        not agree and neither of them has a dimension equal to 1. Therefore, a value error is raised.
        -------------------------------------------------------------------------------------------
        
        Test for numpy array broadcasting Rule 3.
        "If in any dimension the sizes disagree and neither is equal to 1, 
        an error is raised."
        
        Checks if ValueError is raised when dimension sizes disagree and neither
        dimension is equal to 1.
        """
        a = np.ones((2, 3))
        b = np.ones((3, 3))
        
        with self.assertRaises(ValueError):
            c = a + b
````
